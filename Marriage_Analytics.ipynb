<h4>Selecting Multiple rows by label</h4><br>
<p>The df.loc property takes to arguments, the first is the indexes of the slices, since was left open, i.e, df.loc[ : ], there isn’t any row slicing. The second argument is the names of the columns we desire to slice.</p>

```python
selected_data = data.loc[:,['Area Name','Total/ Rural/ Urban', 'Age at marriage',
       'Number of ever married persons - Males',
       'Number of ever married persons - Females']]
```

```python
selected_data
```

```python
#used where clause to slice data

india_data = selected_data.where(selected_data["Total/ Rural/ Urban"]=="Total" )
```

```python
# dropped all null values 
india_data=india_data.dropna()  
```

<h4>Now we need to add column depicting percentages</h4>

```python
india_data['Percentage of Males'] = (india_data["Number of ever married persons - Males"]*100)/242537413.0
india_data['Percentage of Females'] = (india_data["Number of ever married persons - Females"]*100)/273405276.0
```
